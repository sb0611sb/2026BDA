```python
from matplotlib import figure
import koreanize_matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from  pydataset import data
%config InlineBackend.figure_format = 'retina'
plt.rcParams['figure.figsize']=(5,3)
```

```python
tg = data('ToothGrowth')

sns.scatterplot(data=tg, x='dose', y='len', hue='supp')
plt.title('correlation between dose and length')
plt.xlabel('dose')
plt.ylabel('len')
plt.show()
```

```python
sns.boxplot(data=tg,x='dose',y='len',hue='supp')
sns.stripplot(data=tg,x='dose',y='len',hue='supp',dodge=True,alpha=0.6)
plt.title('tooth growth by dose and supplement')
plt.xlabel('len')
plt.ylabel('dose')
plt.legend(title='supp')
plt.tight_layout()
plt.show()
```

```python
vc = tg[tg.supp=='VC'][['len','dose']]
oj = tg[tg.supp=='OJ'][['len','dose']]

plt.plot(vc.dose, vc.len, color='cyan',marker='o',lw=.7)
plt.plot(oj.dose, oj.len, color='deeppink',marker='^',lw=.7, alpha=.5)
plt.title('tooth growth by dose and supplement')
plt.xlabel('dose')
plt.ylabel('length')
plt.legend(['VC','OJ'],loc='lower right')
plt.show()
```

```python
iris = data('iris')
```

```python
n,bins,patches = plt.hist(iris['Sepal.Length'],bins=20,color='yellowgreen',edgecolor='black')
print(n)
print(bins)
print(patches)
plt.show()
```

```python
n,bins,patches = plt.hist(iris['Sepal.Length'],bins=20,color='yellowgreen',edgecolor='white')
for i in range(len(patches)):
  patches[i].set_facecolor(plt.cm.viridis(n[i]/max(n)))

plt.show()
```

```python
sns.histplot(data=iris, x=iris['Sepal.Width'],color='white')
sns.rugplot(data=iris, x=iris['Sepal.Width'])
```

```python
t = [str(i)+'cylinder' for i in mtc.cyl.value_counts().sort_index().index]

plt.pie(mtc.cyl.value_counts().sort_index(),colors=['yellowgreen','lightskyblue','lightcoral'],autopct='%1.1f%%',labels=t)

plt.show()
```

```python
l = [str(i)+'gear' for i in mtc.gear.value_counts().sort_index().index]
c = sns.color_palette('Accent',len(l))

plt.pie(mtc.gear.value_counts().sort_index(),colors=c,autopct='%1.1f%%',labels=l,explode=[0,.1,0],shadow=True,startangle=90)
plt.show()
```

```python
ci = iris[iris.columns[:-1].to_list()].corr()
```

```python
sns.heatmap(ci,annot=True)
plt.show()
```

```python
mask = np.zeros_like(ci, dtype=bool)
mask[np.triu_indices_from(mask)] = True

sns.heatmap(ci,annot=True,mask=mask,cmap='coolwarm',annot_kws={'size':10},vmin=-1,vmax=1,cbar={'shrink':.4},linewidths=.5)
plt.show()
```
