```python
np.random.seed(2025)
n=80
x=np.random.rand(n)
y=np.random.rand(n)
color=np.random.rand(n)
area=(20*np.random.rand(n))**2
plt.scatter(x,y,s=area,c=color,alpha=0.7)
plt.colorbar()
plt.show()
```

```python
x=np.random.randint(-10,10,10)
print(x)
print(x>0)
print(x[x>0])
```

```python
x=np.random.randint(-100,100,100)
y=np.random.randint(-100,100,100)

m1=abs(x)>50
m2=abs(y)>50
x=x[m1+m2]
y=y[m1+m2]

plt.scatter(x,y,c=x,cmap='jet',alpha=0.4)
plt.colorbar()
plt.show()
```

```python
fig, ax=plt.subplots(2,3,figsize=(7,4))
fig.set_size_inches(6,3)

ax[0,1].plot([1,4,9,16])
ax[1,0].plot([1,2,3,4])
ax[1,2].plot([6,5,2,1])
fig.suptitle('제목',fontsize=12,fontweight='bold')
ax[0,0].set_title('(0,0) 빈 플롯')
ax[0,1].set_title('(0,1) 제곱수')
ax[0,2].set_title('(0,2) 빈 플롯')
ax[1,0].set_title('(1,0) 선형')
ax[1,1].set_title('(1,1) 빈 플롯')
ax[1,2].set_title('(1,2) 감소')

plt.tight_layout()
plt.show()
```
